```python
#problem 1
def eng(likes):
  tot=sum(likes)
  print("total likes:",tot)
  if tot>=1000:
    print("Post status: Viral post")
  else:
    print("Post status: Normal engagement")
likes = [200, 150, 300, 400]
eng(likes)
```

```python
#problem 3
data = [60, 75, 80, 70, 75]
def rain(data):
  req= 70
  av=sum(data)/len(data)
  print("Average Rainfall:",av)
  if av>=req:
    print("Adequate Rainfall")
  else:
    print("Inadequate Rainfall")
rain(data)
```

```python
#problem 4
u = ["ravi123","anita_k","kiran007","ravi123","meena90"]
def dup(x):
  dic={}
  for i in x:
    if i in dic:
      dic[i]+=1
    else:
      dic[i]=1
  for j,m in dic.items():
    if m>1:
      print("Duplicate Accounts Found:Yes")
dup(u)
```

```python
#problem 10
def el(units):
    bill = 0
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7
    if bill < 500:
        status = "Low Usage"
    elif 500 <= bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    return bill, status
units_consumed = 250
total_bill, usage_status = el(units_consumed)
print("Units Consumed:", units_consumed)
print("Total Bill: â‚¹", total_bill)
print("Usage Status:", usage_status)
```
