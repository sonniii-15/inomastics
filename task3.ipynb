```python
#problem 1
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
s=max(employees.values())
for name,emp in employees.items():
  if s==emp:
    print("Top performers eligible for bonus:",name,emp)
```

```python
#problem 2
s="Buy mobile phone buy phone online"
o=s.lower()
x=o.split()
l=['buy','phone']
dic={}
for i in x:
  if i in l:
    if i in dic:
      dic[i]+=1
    else:
      dic[i]=1
print(dic)
```

```python
#Problem 3
s= [3, 4, 7, 8, 10, 12, 5]
p=[]
for i in s:
  if i%2==0:
    p.append((s.index(i),i))
print(p)
```

```python
#problem 4
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
dic={}
for i in emails:
  p=i.split("@")[1]
  if p not in dic:
    dic[p]=1
  else:
    dic[p]+=1
print(dic)
```

```python
#problem 5
sales = [1200, 1500, 900, 2200, 1400, 3000]
avg=sum(sales)/len(sales)
surge=avg+(0.3*avg)
for i in sales:
  if i>surge:
    print(f"day{sales.index(i)+1}:{i}")
#day 4 wont be included because it isnt more than the surge
```

```python
#problem 6
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
dic={}
for i in user_ids:
  if i in dic:
    dic[i]+=1
  else:
    dic[i]=1
for c,j in dic.items():
  if j>1:
    print(f"{c}-->{j} times")
```
