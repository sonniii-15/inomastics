```python
#Problem 1: Unique Words in a Sentence
s="python is easy and python is powerful"
m=s.split()
n=[]
for i in m:
  if i not in n:
    n.append(i)
print("Unique word count:",len(n))
print("Unique words:",n)

#Problem 2: Highest salary employee data
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
sal=0
name=[]
for i,j in employees.items():
  if j>sal:
    sal=j
    name=i
print(f"Highest salary:{name}-{sal}")

#Problem 4: Count Products Above a price threshold
numbers = [45, 22, 89, 10, 66]
max=0
min=0
for i in numbers:
  if i>max:
    max=i
  if i>0:
    min=i
    for j in numbers:
      if j<min:
        min=j
print("List:",numbers)
print("Max:",max)
print("Min:",min)

#problem 5 calculate the attendance percentage
attendance = ["P", "P", "A", "P", "P"]
count=0
for i in attendance:
  if i=="P":
    count+=1
per=(count/len(attendance))*100
print(f"Attendance perecentage is: {per}%")

#Problem 6 remove duplicate phone numbers
phone_numbers = [9876543210, 9123456789, 9876543210]
p=set(phone_numbers)
print("Unique phone numbers:",p)

#program 7 count the character frequency
s="i love python programming"
p=s.split()
d={}
for i in p:
  for j in i:
    if j in d:
      d[j]+=1
    else:
      d[j]=1
print(d)

#problem 8 Convert list to tuple
l=[10,20,30]
m=tuple(l)
print("Tuple:",m)

#problem 9 Check if a key exists in a dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
i="Ravi"
if i in employees:
  print("Employee exists")
else:
  print("Employee does not exist")

#Program 10 calculate average marks
l=[90,92,95,96,93]
sum=0
for i in l:
  sum+=i
avd=sum/len(l)
print("Average marks:",avd)
```
